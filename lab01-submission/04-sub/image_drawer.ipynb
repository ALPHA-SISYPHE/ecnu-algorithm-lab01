```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sort_time_comparison.csv")

plt.figure(figsize=(13, 8))

plt.plot(df["scale"], df["QuickSort_ms"],   "o-", label="QuickSort")
plt.plot(df["scale"], df["MergeSort_ms"], "s-", label="MergeSort")

plt.xscale("log")
plt.yscale("log")

plt.xlabel("data scale n")
plt.ylabel("time (ms)")
plt.title("QuickSort vs MergeSort")
plt.grid(True, which="both", ls="--")
plt.legend()

plt.tight_layout()
plt.show()
```
